Notebook to allow viewing of catalogue entry natural language fields. The raw data is used for this.

```python
from pathlib import Path
import pandas as pd


# location of modules.csv - get path of current working directory, then go up one level, then navigate to modules.csv in ./data/
modules_path = Path.cwd().parent.joinpath('data/raw/modules.csv')

# load modules.csv, semi-colon is the separator used in the file 
modules = pd.read_csv(modules_path, sep = ';')

pd.set_option('display.max_colwidth', None)
```

In the below, assign `module_code_a` and `module_code_b` to the two module codes you wish to look at. Then, run the cell; the output will show the catalogue text for the two modules.

```python
module_code_a = 'MCD8007'
module_code_b = 'HIS3351'

# get the text fields for the given ModuleCodes
modules_view = modules[(modules['ModuleCode'] == module_code_a) | (modules['ModuleCode'] == module_code_b)][['ModuleCode',
                                                                                                             'Aims',
                                                                                                             'OutlineOfSyllabus',
                                                                                                             'IntendedKnowledgeOutcomes',
                                                                                                             'IntendedSkillOutcomes']]

# left align the text for easier readability
modules_view = modules_view.style.set_properties(**{'text-align': 'left'})
modules_view
```
